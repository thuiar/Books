```python
import csv
import random
import numpy as np
import os

with open('td.tsv', "r", encoding="utf-8") as f:
    reader = csv.reader(f,delimiter = "\t")
    texts = []
    labels = []
    for row in reader:
        text = row[0]
        label = row[1]
        if text == 'text':
            continue
        texts.append(text)
        labels.append(label)

num_all = len(labels)
label_list = list(np.unique(np.array(labels)))
labels = np.array(labels)
res_ids = []
for label in label_list:
    num = int(len(labels[labels == label]) / 2)
    pos = list(np.where(labels == label)[0])
    res_ids.extend(random.sample(pos, num))
    
random.shuffle(res_ids)

res_texts = []
res_labels = []
for idx in res_ids:
    res_texts.append(texts[idx])
    res_labels.append(labels[idx])
    

data_path = 'ori_dev.tsv'
with open(data_path, "w", encoding="utf-8") as f:
    writer = csv.writer(f, delimiter="\t")
    writer.writerow(['text', 'label'])
    for text, label in zip(res_texts, res_labels):
        writer.writerow([text, label])   

test_texts = []
test_labels = []
test_ids = []
for idx in range(num_all):
    if idx not in res_ids:
        test_ids.append(idx)
        
random.shuffle(test_ids)
for idx in test_ids:
    test_texts.append(texts[idx])
    test_labels.append(labels[idx])
    
data_path = 'ori_test.tsv'
with open(data_path, "w", encoding="utf-8") as f:
    writer = csv.writer(f, delimiter="\t")
    writer.writerow(['text', 'label'])
    for text, label in zip(test_texts, test_labels):
        writer.writerow([text, label])   
```
